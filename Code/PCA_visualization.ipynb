# Eigenhands for recognition with PCA

```python
# Import libraries
from matplotlib import pyplot as plt
from matplotlib.image import imread
import numpy as np
import os
```

```python
import matplotlib.font_manager
from IPython.core.display import HTML

def make_html(fontname):
    return "<p>{font}: <span style='font-family:{font}; font-size: 24px;'>{font}</p>".format(font=fontname)

code = "\n".join([make_html(font) for font in sorted(set([f.name for f in matplotlib.font_manager.fontManager.ttflist]))])

HTML("<div style='column-count: 2;'>{}</div>".format(code))
```

```python
dataset_path = 'C:/Users/gabri/OneDrive/Documentos/GitHub/Hand_gesture_recognition/Own_dataset_hands/'
dataset_dir  = os.listdir(dataset_path)
width  = 576
height = 400

names = ['Palm', 'C', 'Fist', 'Ok', 'Peace', 'I Love you', 'L']
```

```python
print('Train Images:')
Gestures = []
for g in range(7):
    Gestures.append('Gesture' + str(g) + '_frame00.png')
Gestures_tensor   = np.ndarray(shape=(len(Gestures), height*width), dtype=np.float64)

fig = plt.figure(figsize=(15, 7))
fig.suptitle('Gestos para el set de entrenamiento', fontsize=26, fontname='Times New Roman')

for i in range(len(Gestures)):
    img = plt.imread(dataset_path + Gestures[i])
    Gestures_tensor[i,:] = np.array(img, dtype='float64').flatten()
    plt.subplot(2,4,1+i)
    plt.imshow(img, cmap='gray')
    plt.title(str(i+1)+') '+str(names[i]), fontsize=15, fontname='Times New Roman')
    #plt.tick_params(labelleft='off', labelbottom='off', bottom='off',top='off',right='off',left='off', which='both')
plt.show()
```

```python
dataset_path2 = 'C:/Users/gabri/OneDrive/Documentos/GitHub/Hand_gesture_recognition/Images/'
dataset_dir2  = os.listdir(dataset_path2)
width2  = 200
height2 = 200

average = plt.imread(dataset_path2 + 'average_hand.png')
plt.plot()
plt.imshow(average, cmap='gray')
plt.show()
```

```python
print('Eigenhands:')
Eigenhands = []
for e in range(10):
    Eigenhands.append('Eigenhand '+str(e)+'.png')
eigenhand_tensor   = np.ndarray(shape=(len(Eigenhands), height2*width2), dtype=np.float64)

fig2 = plt.figure(figsize=(15, 7))
fig2.suptitle('Primeras 10 eigenhands obtenidas', fontsize=26, fontname='Times New Roman')

for i in range(len(Eigenhands)):
    img = plt.imread(dataset_path2 + Eigenhands[i])
    eigenhand_tensor[i,:] = np.array(img, dtype='float64').flatten()
    plt.subplot(2,5,1+i)
    plt.imshow(img, cmap='gray')
    #plt.title('Título', fontsize=15, fontname='Times New Roman')
    #plt.tick_params(labelleft='off', labelbottom='off', bottom='off',top='off',right='off',left='off', which='both')
plt.show()
```
